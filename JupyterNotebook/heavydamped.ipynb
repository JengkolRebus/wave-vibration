```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Menentukan nilai parameter dalam persamaan GHS teredam berat.
Parameter: 
- t = waktu
- G = Amplitudo
- r = resistan
- s = gaya hambat
- m = massa

```python
t = np.arange(0.0, 1.5, 0.01)   # waktu getaran selama 1.5 s. perubahan waktu dari 0 s menuju 1.5 s bertambah sebesar 0.01.
G = 1
r = (20, 25, 30)    # 3 resistansi berbeda (r1, r2, r3)
m = 2
s = 2
```

```python
fig, ax = plt.subplots()    # mempersiapkan canvas untuk plotting data posisi yang akan ditampilkan

for i in r:
    p = (-i*t)/(2*m)
    q = (np.square(i)/(4*np.square(m)))-(s/m)
    x = G * np.exp (p) * np.sinh(np.sqrt(q)) * t
    leg = 'r = ' + str(i)
    ax.plot(t, x, label=leg)
    ax.legend()

ax.set(xlabel='t', ylabel='x', 
        title='x = G exp(-rt/2m) sinh((r^2/4m^2)-(s/m))^1/2 t''\n'
        'G=1, s=2, m=2')
ax.grid()
plt.show()
```
